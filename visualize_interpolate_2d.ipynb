```python
import matplotlib.pyplot as plt
%matplotlib inline
from IPython import display

import os
ROOT_DIR = '.'
import numpy as np
import cPickle
```

```python
# settings
log_file = os.path.join(ROOT_DIR, 'cache', 'cifar10/resnet8_sgd-0300_resnet8_adam-0300_linear_interpolate.pkl')
```

```python
with open(log_file, 'rb') as fd:
    log = cPickle.load(fd)

alpha_list = log['alpha_list']
train_error_list = log['train_error_list']
val_error_list = log['val_error_list']
```

```python
plt.figure(figsize=(8, 6))
plt.title('interpolated error surface')
plt.xlim(0, 1.1)
plt.ylim(0, 1)
plt.xlabel('alpha')
plt.ylabel('error')
if train_error_list is not None and len(train_error_list) > 0:
    plt.plot(alpha_list, train_error_list, c='b', label='train')
    plt.scatter(alpha_list, train_error_list, c='b')

if val_error_list is not None and len(val_error_list) > 0:
    plt.plot(alpha_list, val_error_list, c='g', label='val')
    plt.scatter(alpha_list, val_error_list, c='g')

plt.legend(loc=1)
plt.show()
```
