```python
import matplotlib.pyplot as plt
%matplotlib inline
from IPython import display

import os
ROOT_DIR = '.'
import numpy as np
import cPickle
```

```python
# settings
log_file = os.path.join(ROOT_DIR, 'cache', 'resnet_sgd-0000_resnet_sgd-0300_linear_interpolate.pkl')
```

```python
with open(log_file, 'rb') as fd:
    log = cPickle.load(fd)

alpha_list = log['alpha_list']
train_error_list = log['train_error_list']
val_error_list = log['val_error_list']
```

```python
plt.figure(figsize=(8, 6))
plt.title('interpolated error surface')
plt.xlabel('alpha')
plt.ylabel('error')
plt.plot(alpha_list, train_error_list, c='b', label='train')
plt.scatter(alpha_list, train_error_list, c='b')
plt.plot(alpha_list, val_error_list, c='g', label='val')
plt.scatter(alpha_list, val_error_list, c='g')
plt.legend(loc=1)
plt.show()
```
